```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches

from matplotlib.path import Path

plt.ion()
matplotlib.is_interactive()
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3],[2,4,6])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3],[2,4,6])

ax.annotate("min value", xy=(1,2), xytext=(1,3), arrowprops=dict(color="g"))

plt.show()
```

```python
import numpy as np
```

```python
np.random.seed(0)

x1 = -1 + np.random.randn(100)
y1 = -1 + np.random.randn(100)

x2 = -1 + np.random.randn(100)
y2 = -1 + np.random.randn(100)
```

```python
fig, ax = plt.subplots()
ax.scatter(x1, y1, color="r")
ax.scatter(x2, y2, color="g")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x1, y1, color="r")
ax.scatter(x2, y2, color="g")

bbox_props = dict(boxstyle="square", facecolor="w", alpha=0.5)

ax.text(-3, -3, "Sample A", ha="center", va="center", size=20, bbox=bbox_props)
ax.text(0, 0, "Sample B", ha="center", va="center", size=20, bbox=bbox_props)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x1, y1, color="r")
ax.scatter(x2, y2, color="g")

bbox_props = dict(boxstyle="square", facecolor="w", alpha=0.5)

ax.text(-3, -3, "Sample A", ha="center", va="center", size=20, bbox=bbox_props)
ax.text(0, 0, "Sample B", ha="center", va="center", size=20, bbox=bbox_props)

arrow_bbox_props = dict(boxstyle="rarrow", facecolor="#EBF5FB", edgecolor="b", linewidth=2, alpha=0.7)

ax.text(-1, -3, "Direction", ha="center", rotation=60, size=15, bbox=arrow_bbox_props)

plt.show()
```
