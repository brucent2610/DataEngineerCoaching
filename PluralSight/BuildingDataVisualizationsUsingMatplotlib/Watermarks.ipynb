```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.ion()
matplotlib.is_interactive()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [2, 4, 6, 8])
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [2, 4, 6, 8])
ax.text(1, 4, "Do not distribute",
       fontsize=30,
       color="red",
       ha="left",
       va="bottom",
       alpha=0.5)
```

```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(221)
ax1.plot([1, 2, 3, 4],
        [2, 4, 6, 8])
ax1.set_label("Straight Line")
ax1.text(1, 4, "Do not distribute",
       fontsize=20,
       color="red",
       ha="left",
       va="bottom",
       alpha=0.5)

ax2 = fig.add_subplot(222)
ax2.plot(x, np.sin(x))

ax3 = fig.add_subplot(223)
ax3.plot(x, np.cos(x))

fig.text(0.1, 0.5, "Propery of Pluralsight",fontsize=40,
       color="gray",
       ha="left",
       va="bottom",
       alpha=0.3)
```
