```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.ion()
matplotlib.is_interactive()
```

```python
x = np.linspace(start=0, stop=10, num=50)
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x), label="sine curve", color="green")
plt.plot(x, np.cos(x), label="cosine curve", color="m")

plt.title("Playing with Plots")
```

```python
random_array = np.random.randn(20)
random_array
```

```python
plt.plot(random_array, color="green")
```

```python
plt.plot(random_array, color="green", linestyle=":")
```

```python
plt.plot(random_array, color="green", linestyle="--", linewidth=3)
```

```python
plt.plot(random_array, color="green", marker="d")
```

```python
plt.plot(random_array, color="green", marker="d", markersize=10)
```

```python
plt.plot(random_array, color="green", marker="d", linestyle="None")
```
