```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.ion()
matplotlib.is_interactive()
```

```python
products = pd.read_csv("./DataCollected/Products.csv", delimiter=';', low_memory=False)
```

```python
products.head()
```

```python
brand_sum = products['brand'].value_counts()
brand_sum
```

# Các hãng đang cung cấp Card đồ họa, số lượng sản phẩm của mỗi hãng.

```python
brand_sum.head(20).plot(kind='bar')
```

# Phân bố giá của các sản phẩm (Mức giá phổ biến là bao nhiêu)

```python
price_sum = products['price'].value_counts()
price_sum
```

```python
price_sum.head(20).plot(kind='bar')
```

# Phân bố giá sản phẩm theo hãng

```python
brands = brand_sum.head(10).index.unique()
brands
```

```python
plt.figure(figsize=(20,12))
for i in brands:
    price_by_brand = products[products["brand"] == i]["price"]
    x = price_by_brand.index.tolist()
    y = price_by_brand.tolist()
    plt.plot(x, y, label=i)
plt.legend()
plt.show()
```

# Biểu diễn mối liên hệ giữa giá sản phẩm và rating của người dùng

```python
plt.figure(figsize=(20,12))
plt.scatter(products['rating'], products['price'], c ="red") 
plt.show()
```
